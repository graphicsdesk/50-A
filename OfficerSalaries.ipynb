# Payroll Records of Substantiated Cops
#### The payroll records of NYPD officers with substantiated complaints
#### The top table has average values calculated from 2014 to 2019
#### Individual tables for each officer (separated by every fiscal year) will follow
### Definitions(?):
baseSalary (in dollars) - The "base salary" assigned to the officer<br>
regHours - Number of hours worked by the officer<br>
regGrossPay (in dollars) - Actual amount of money received by officer as their "base salary" (May differ from assigned base salary)<br>
overtimeHours - Number of overtime hours worked by the officer<br>
overtimePay (in dollars) - Amount received by the officer as payment for working overtime<br>
otherPay (in dollars) - Amount received by the officer that cannot be classified under their base salary or overtime pay<br>
%changeBaseSalary - Percent Change in base salary compared to the previous year

```python
#html reader
htmlList = ["yuseffhamm","kierancreighton","rolandocarbonell","wilbertrodriguez","rafaeliturralde","michaelscoloveno","marcyvelez","robertcattani","ronniemorales","welingtonfeliz","benjamingottlieb","barrystokes","freddytavares","johnandryuk","eleodoromata","edsonbrutus","christophertorturo","marydaniellesessomsswinea","alulahaileselassie"]
iL = [];
stringList = []
for z in htmlList:
    with open(r'html/'+z+r'.html','r') as f:
        stringList.append(f.read())
```

```python
#make individual tables
import pandas as pd
import re
iL=[]
fy=re.compile("""Year <\/span>FY(.*?)<\/div>""")
sal=re.compile("""Salary <\/span>(.*?)<\/div>""")
regHrs=re.compile("""Regular Hours <\/span>(.*?)<\/div>""")
grossPay=re.compile("""Gross Pay <\/span>(.*?)<\/div>""")
othrPay=re.compile("""Other Pay <\/span>(.*?)<\/div>""")
otPay=re.compile("""Overtime <\/span>(.*?)<\/div>""")
otHrs=re.compile("""Overtime Hours <\/span>(.*?)<\/div>""")
regexList = [fy, sal, regHrs, grossPay, otHrs, otPay, othrPay ]
for y in (stringList):
    arr=[]
    for x in regexList:
        tmpList=x.findall(y)
        tmpList.reverse()
        arr.append(tmpList)
    df = pd.DataFrame(arr, ['Year','baseSalary','regHours','regGrossPay','overtimeHours','overtimePay','otherPay'])
    iL.append(df.transpose())
hamm=iL[0];creighton=iL[1];rodriguez=iL[2];carbonell=iL[3];iturralde=iL[4];scoloveno=iL[5];velez=iL[6];cattani=iL[7];morales=iL[8];feliz=iL[9];gottlieb=iL[10];stokes=iL[11];tavares=iL[12];andryuk=iL[13];mata=iL[14];brutus=iL[15];torturo=iL[16];sessomsswinea=iL[17];selassie=iL[18];
```

```python
#make overall table
firstNames = ['YUSEFF', "KIERAN", "ROLANDO", "WILBERT", "RAFAEL", "MICHAEL", "MARCY", "ROBERT", "RONNIE", "WELINGTON", "BENJAMIN", "BARRY", "FREDDY", "JOHN", "ELEODORO", "EDSON", "CHRISTOPHER", "MARY-DANIELLE", "ALULA"]
lastNames = ['HAMM', "CREIGHTON", "CARBONELL", "RODRIGUEZ JR.", "ITURRALDE JR.", "SCOLOVENO", "VELEZ", "CATTANI", "MORALES", "FELIZ", "GOTTLIEB", "STOKES", "TAVARES", "ANDRYUK", "MATA", "BRUTUS", "TORTURO", "SESSOMS-SWINEA", "HAILE SELASSIE"]
avgList=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
for y in range(len(stringList)):
    avgList[y].append(lastNames[y])
    avgList[y].append(firstNames[y])
    for x in (range(len(regexList)-1)):
        brr = regexList[x+1].findall(stringList[y])
        tempnum=0
        for z in range(len(brr)):
            if(brr[z][-3:]=="hrs"):
                tempnum+=int(brr[z][:-4].replace(",",""))
            else:
                tempnum+=int(brr[z].replace(",",""))
        avgList[y].append(tempnum/len(brr))
copSalaries2 = pd.DataFrame(data=avgList,columns=['Last Name','First Name','averageBaseSalary','averageRegularHours','averageRegularGrossPay','averageOvertimeHours','averageOvertimePay','averageOtherPay'])
#display(copSalaries2)
```
